```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
Data = pd.read_csv('grade1.csv')
Data.corr()*100
```

```python
X = Data[['Calculus','English','ENGCC304']]
y = []
for i in Data['ENGCE101']:
    if(i > 0):
        y.append(1)
    else:
        y.append(0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
predictModel = LogisticRegression()
predictModel.fit(X_train,y_train)
y_pred = predictModel.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)

predictModel.predict_proba([[3,0,2]])
```
